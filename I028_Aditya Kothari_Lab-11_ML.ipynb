```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn import preprocessing
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
df['Class']=label_encoder.fit_transform(df['Class'])
df['Gender']=label_encoder.fit_transform(df['Gender'])
df['Car Type']=label_encoder.fit_transform(df['Car Type'])
df['Shirt Size']=label_encoder.fit_transform(df['Shirt Size'])
```

```python
x = df.drop(['Class','Customer ID'],axis=1)
y = df['Class']
```

```python
clf_gini = DecisionTreeClassifier( criterion = 'gini', random_state = 100 , max_depth = None , min_samples_leaf = 1)
```

```python
tree.plot_tree(clf_gini)
```

```python
clf_entropy.fit(x,y)
```

```python
tree.plot_tree(clf_entropy)
```

```python
X = df1.drop(['Outcome'] , axis = 1)
Y = df1['Outcome']
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=100)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
clf_gini1 = DecisionTreeClassifier( criterion = 'gini', random_state = 100 , max_depth = 5 , min_samples_leaf = 1)
```

```python
clf_gini1.fit(X_train,Y_train)
```

```python
tree.plot_tree(clf_gini1)
```

```python
clf_entropy1 = DecisionTreeClassifier( criterion = 'entropy', random_state = 100 , max_depth = 5 , min_samples_leaf = 1)
```

```python
clf_entropy1.fit(X_train,Y_train)
```

```python
tree.plot_tree(clf_entropy1)
```

```python
pred=clf_gini1.predict(X_test)
pred
```

```python
cm = confusion_matrix(Y_test, pred)
ac = accuracy_score(Y_test,pred)
print(cm)
print(ac)
```

```python
pred1=clf_entropy1.predict(X_test)
pred1
```

```python
cm1 = confusion_matrix(Y_test, pred1)
ac1 = accuracy_score(Y_test,pred1)
print(cm1)
print(ac1)
```

The decision tree based on gini index performs better than the entropy based tree.

Decreasing the maximum depth of the tree increases it's accuracy because it reduces overfitting.

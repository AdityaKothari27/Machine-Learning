```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import cv2
```

```python
dataset,classes = make_blobs(n_samples=200,n_features=2,centers=4,random_state=0,cluster_std=0.50)
```

```python
df=pd.DataFrame(dataset,columns=['var1','var2'])
```

```python
df.head()
```

```python
df.shape
```

```python
plt.scatter(df['var1'],df['var2'])
```

```python
kmeans = KMeans(4)
```

```python
kmeans.fit(df)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
four_clusters = kmeans.fit_predict(df)
df_with_clusters = df.copy()
df_with_clusters['clusters'] = four_clusters
```

```python
df_with_clusters.head()
```

```python
plt.scatter(df_with_clusters['var1'],df_with_clusters['var2'],c=df_with_clusters['clusters'],cmap='gray')
```

```python
opti = []
for i in range(1,10):
  kmeans = KMeans(i)
  kmeans.fit(df)
  opti_cluster = kmeans.inertia_
  opti.append(opti_cluster)
```

```python
opti
```

```python
number_clusters = range(1,10)
plt.plot(number_clusters,opti)
plt.title("The Elbow Curve")
plt.xlabel("Number of clusters")
plt.ylabel("Opti")
```

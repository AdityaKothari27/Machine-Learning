```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN 
from sklearn.datasets import make_circles
from sklearn.datasets import make_moons
```

```python
dataset,classes = make_blobs(n_samples=250,n_features=2,centers=1,random_state=1,cluster_std=0.3)
```

```python
plt.scatter(dataset[:,0],dataset[:,1])
```

```python
classes
```

```python
dbscan = DBSCAN(eps = 0.3, min_samples = 20)
```

```python
pred = dbscan.fit_predict(dataset)
```

```python
pred
```

```python
outlier_index = np.where(pred==-1)

outlier_val = dataset[outlier_index]
```

```python
outlier_index
```

```python
outlier_val
```

```python
plt.scatter(dataset[:,0],dataset[:,1])
plt.scatter(outlier_val[:,0],outlier_val[:,1],color = 'r')
```

Decreasing the eps value, increases the no. of outliers.

Decreasing the value of minimum samples, decreases the no. of outliers.
